# Here import the libraries to need recommnedenation system

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf
```

# **Feature Extraction**

```python
# Convert genres to numerical form using TfidfVectorizer
df['genre']=df['genre'].fillna('')
tfidf_matrix=tfidf.fit_transform(df['genre'])
```

```python
# Normalize rating and combine it with genre data (optional
df_features=pd.DataFrame(tfidf_matrix.toarray())
df_features['rating']=df['rating']
```

```python
df_features
```

# **Recommendation System:**

```python
df_sim=1-pairwise_distances(df_features.values,metric='cosine')
df_sim
```

```python
df_similar=pd.DataFrame(df_sim)
df_similar
```

```python
# Function to recommend anime

def recommend_anime(anime_title, cosine_sim=df_sim, df=df):
    indices = pd.Series(df.index, index=df['name']).drop_duplicates()
    idx = indices[anime_title]

    sim_scores=list(enumerate(df_sim[idx]))
    sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
    sim_scores=sim_scores[1:11]
    df_indices=[i[0] for i in sim_scores]
    return df['name'].iloc[df_indices]
print(recommend_anime('Naruto'))
```
