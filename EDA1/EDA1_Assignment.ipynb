# Detect and treat outliers if necessary.

```python
Q1 = df.quantile(0.25)
print('Q1 :-',Q1)
```

```python
Q3 = df.quantile(0.75)
print('Q3 :-',Q3)
```

```python
IQR = Q3 - Q1
print('IQR :-',IQR)
```

```python
LL = Q1 - 1.5 * IQR
print('Lower Limit :-',LL)
```

```python
UL = Q3 + 1.5 * IQR
print('Lower Limit :-',UL)
```

```python
outliers = ((df < (LL)) | (df > (UL)))
print('Outliers :-', outliers)
```

# **3.	Data Visualization:**

# Create histograms or boxplots to visualize the distributions of various numerical variables.

# **HISTOGRAM**

# Histograms for the distribution of numerical variables

```python
plt.figure(figsize=(15, 10))
df.hist(bins=15, color='blue', edgecolor='black')
plt.suptitle("Histograms of Cardiographic Data", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(5, 6))

for col in df.columns:
    plt.hist(df[col], bins=15, color='blue', edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.show()

plt.suptitle("Histograms of Cardiographic Data", fontsize=16)
```

# Boxplots to detect outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplots for All Variables")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5, 6))
for col in df.columns:
  sns.boxplot(data=df[col])
  plt.title("Boxplots for All Variables")
  plt.xticks(rotation=90)
  plt.show()
```

# Use bar charts or pie charts to display the frequency of categories for categorical variables.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='NSP', data=df)
plt.title('Bar Chart of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

# Assuming 'HealthStatus' is a categorical column in the dataset

# Categories: ['Normal', 'Warning', 'Critical']

# **BAR CHARTS**

```python
df['HealthStatus'] = pd.cut(df['LB'], bins=[100, 120, 140, 160], labels=['Normal', 'Warning', 'Critical'])

# Step 2: Bar Chart for frequency of categories
plt.figure(figsize=(8, 6))
df['HealthStatus'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Bar Chart of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

# ***PIE CHARTS ***

```python
plt.figure(figsize=(8, 6))
df['HealthStatus'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'orange', 'red'])
plt.title('Pie Chart of Health Status')
plt.show()
```

# Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['LB'], df['ASTV'], alpha=0.5)
plt.title('Scatter Plot between LB and ASTV')
plt.xlabel('LB (Baseline Fetal Heart Rate)')
plt.ylabel('ASTV (Abnormal Short Term Variability)')
plt.show()
```

# Correlation Heatmap

```python
# Convert 'HealthStatus' column to numerical representation using one-hot encoding
df = pd.get_dummies(df, columns=['HealthStatus'])

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Cardiographic Data")
plt.show()
```
