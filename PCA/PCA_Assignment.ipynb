# **Task 2: Dimensionality Reduction with PCA:**

# Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df)
df_scaled[:5]
```

# Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pca.fit(df_scaled)
```

# SCREE PLOT

```python
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cummulative Explaned variance')
plt.show()
```

```python
pca=PCA(n_components=2)
df_pca=pca.fit_transform(df_scaled)
df_pca[:5]
```

# **Task 3: Clustering with Original Data**

# Apply a clustering algorithm (e.g., K-means) to the original dataset.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans=KMeans(n_clusters=3, random_state=42)
labels=kmeans.fit_predict(df_scaled)
```

```python
# Silhouette Score
sil_score = silhouette_score(df_scaled, labels)
print('Silhouette Score of original data :-',sil_score)
```

# Visualize the clustering results using appropriate plots.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df_scaled[:,0],y=df_scaled[:,1],hue=labels,palette='viridis')
plt.title('Clustering of original data')
plt.show()
```

# **Task 4: Clustering with PCA Data:**

# Apply the same clustering algorithm to the PCA-transformed dataset.

```python
# Fit K-Means clustering to PCA -tranformed data
labels_pca=kmeans.fit_predict(df_pca)
labels_pca
```

```python
# Silhouette Score for PCA data

sil_score_pca=silhouette_score(df_pca,labels_pca)
print('Silhpuette Score for PCA data :-',sil_score_pca)
```

# Visualize the clustering results obtained from PCA-transformed data.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df_pca[:,0],y=df_pca[:,1],hue=labels_pca,palette='coolwarm')
plt.title('Clustering on PCA transformed data')
plt.show()
```

# **Task 5: Comparison and Analysis:**

# Compare the clustering results obtained from the original dataset and PCA-transformed data.

```python
print('Silhouette Score of original data :-',sil_score)
print('\n')
print('Silhpuette Score for PCA data :-',sil_score_pca)
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
# Davies-Bouldin score for original data

db_original = davies_bouldin_score(df_scaled, labels)
print(f'Davies-Bouldin Score for Original Data: {db_original}')
```

```python
# Davies-Bouldin score for PCA data

db_pca = davies_bouldin_score(df_pca, labels_pca)
print(f'Davies-Bouldin Score for PCA Data: {db_pca}')
```

```python
# Compare cluster assignments

comparison = np.c_[labels, labels_pca]
print(pd.DataFrame(comparison, columns=["Original Data Labels", "PCA Data Labels"]).head())
```

```python
print("Original Data Cluster Centers:")
print(kmeans.cluster_centers_)
print('\n')
print("PCA Data Cluster Centers:")
print(kmeans.fit(df_pca).cluster_centers_)
```
