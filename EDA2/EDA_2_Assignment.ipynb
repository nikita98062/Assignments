# Handling the missing values

```python
df_cleaned=df.dropna()
```

# Apply scaling techniques to numerical features:

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler_standard=StandardScaler()
numerical_df = df_cleaned.select_dtypes(include=np.number)
df_standard_scaled = pd.DataFrame(scaler_standard.fit_transform(numerical_df), columns=numerical_df.columns)
```

```python
df_standard_scaled
```

# Min-Max Scaling

```python
scaler_minmax = MinMaxScaler()
numerical_df = df_cleaned.select_dtypes(include=np.number)
print(df_cleaned.columns)
data_minmax_scaled = pd.DataFrame(scaler_minmax.fit_transform(numerical_df), columns=numerical_df.columns)
```

# **2. Encoding Techniques:**

# Apply One-Hot Encoding to categorical variables with less than 5 categories.

```python
categorical_df = df_cleaned.select_dtypes(include='category')
categorical_columns=categorical_df.columns
df_one_hot_encoded = pd.get_dummies(df_cleaned, columns=categorical_columns, drop_first=True)
df_one_hot_encoded
```

# **3. Feature Engineering:**

# Create a New Features

```python
df_cleaned['age_group']=pd.cut(df_cleaned['age'],bins=[0,30,60,100],labels=['young','middle','senior'])
```

# Combining marital status

```python
df_cleaned['marital_status_type'] = df_cleaned['marital_status'].apply(lambda x: 'Single' if 'Never-married' in x else 'Coupled')
```

```python
df_cleaned
```

# Log transformation for skewed numerical features

```python
df_cleaned['capital_gain_log']=np.log1p(df_cleaned['capital_gain'])
```

```python
df_cleaned
```

# **4. Feature Selection:**

# Use isolation Forest to Detect outliers

```python
from sklearn.ensemble import IsolationForest
```

# Fit isolation forest

```python
iso_forest=IsolationForest(contamination=0.05)
outliers=iso_forest.fit_predict(df_cleaned.select_dtypes(include=['int64','float64']))
outliers
```

# Remove outliers

```python
df_no_outliers=df_cleaned[outliers==1]
```

```python
df_no_outliers
```
